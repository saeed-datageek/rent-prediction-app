```python
def remove_outliers(df, column='price', low_quantile=0.25, high_quantile=0.75, multiplier=1.5):
    Q1 = df[column].quantile(low_quantile)
    Q3 = df[column].quantile(high_quantile)
    IQR = Q3 - Q1
    low_limit = Q1 - multiplier * IQR
    high_limit = Q3 + multiplier * IQR
    return df.loc[(df[column] >= low_limit) & (df[column] <= high_limit)]
```

```python
class CustomTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, columns=None, values=['−']):
        self.columns = columns if columns is not None else ['beds', 'baths','parking']
        self.values = values
    def fit(self, X, y=None):
        return self
    def transform(self, X,y=None):
        X_copy = X.copy()  # Make a copy of the input DataFrame to avoid modifying the original data
        for col in self.columns:
            X_copy[col] = X_copy[col].apply(lambda x: 0 if x == self.values else x)
        return X_copy
```

```python
def clean_text(text):
    
    # lower case characters only
    text = text.lower() 
    
    # remove urls
    text = re.sub('http\S+', ' ', text)
    
    # only alphabets, spaces and apostrophes 
    text = re.sub("[^a-z' ]+", ' ', text)
    
    # remove all apostrophes which are not used in word contractions
    text = ' ' + text + ' '
    text = re.sub("[^a-z]'|'[^a-z]", ' ', text)
    
    return text.split()

df['address'] = df['address'].apply(lambda x: clean_text(x))

df.sample(5)
```

```python
text_embedding_dimension = 200
embeddings_index_200 = {}
path_to_glove_file = r'C:\Users\Maedeh\Downloads\glove.6B\glove.6B.{}d.txt'.format(text_embedding_dimension)
#with open(r'C:\Users\Maedeh\Downloads\glove.6B\glove.6B.100d.txt', 'r', encoding='utf-8') as f:
with open(path_to_glove_file, 'r', encoding='utf-8') as f:
    for line in f:
        word, coefs = line.split(maxsplit=1)
        coefs = np.fromstring(coefs, "f", sep=" ")
        embeddings_index_200[word] = coefs

print("Found %s word vectors." % len(embeddings_index_200))
```

```python
def text_embed(words):
    
    unknown_indices = []
    mean = np.zeros(text_embedding_dimension)
    
    for i in range(len(words)):
        if words[i] in embeddings_index_200:
            words[i] = embeddings_index_200[ words[i] ]
            mean += words[i]
        else:
            unknown_indices.append(i)
            
    mean /= len(words)-len(unknown_indices)
    
    # unknown words in the text are represented using the mean of the known words
    for i in unknown_indices:
        words[i] = mean
    
    return np.array(words)

df['address'] = df['address'].apply(lambda words: text_embed(words))
```

```python
import numpy as np
import pandas as pd
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense, Input, Concatenate
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load your dataset (replace with your actual data)
df = pd.read_csv('path/to/your/dataset.csv')

# Preprocess the address column (e.g., remove punctuation, lowercase)
df['address'] = df['address'].str.lower()

# Tokenize the address text
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['address'])
sequences = tokenizer.texts_to_sequences(df['address'])

# Pad sequences to a fixed length
max_sequence_length = 20  # Set as needed
X_text = pad_sequences(sequences, maxlen=max_sequence_length)

# Load pre-trained GloVe embeddings (e.g., 100-dimensional vectors)
embedding_dim = 100
embeddings_index = {}
with open('path/to/glove.6B.100d.txt', 'r', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs

# Create an embedding matrix
num_words = len(tokenizer.word_index) + 1
embedding_matrix = np.zeros((num_words, embedding_dim))
for word, i in tokenizer.word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector

# Numeric feature: beds
X_beds = df['beds'].values.reshape(-1, 1)

# Combine address embeddings and beds feature
input_text = Input(shape=(max_sequence_length,))
embedding_layer = Embedding(input_dim=num_words, output_dim=embedding_dim, weights=[embedding_matrix], input_length=max_sequence_length, trainable=False)(input_text)
flatten_layer = Flatten()(embedding_layer)
input_beds = Input(shape=(1,))
concatenated = Concatenate()([flatten_layer, input_beds])
dense_layer = Dense(64, activation='relu')(concatenated)
output_layer = Dense(1)(dense_layer)

model = Model(inputs=[input_text, input_beds], outputs=output_layer)
model.compile(optimizer='adam', loss='mean_squared_error')

# Split data into train and test sets
X_train_text, X_test_text, X_train_beds, X_test_beds, y_train, y_test = train_test_split(X_text, X_beds, df['price'], test_size=0.2, random_state=42)

# Train the model
model.fit([X_train_text, X_train_beds], y_train, epochs=5, batch_size=64, validation_data=([X_test_text, X_test_beds], y_test))

# Evaluate the model
y_pred = model.predict([X_test_text, X_test_beds])
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")

# Make predictions for new addresses and beds
new_address = ["123 Main St, City, State"]
new_sequences = tokenizer.texts_to_sequences(new_address)
new_padded_sequences = pad_sequences(new_sequences, maxlen=max_sequence_length)
new_beds = np.array([[3]])  # Example: 3 beds
predicted_price = model.predict([new_padded_sequences, new_beds])
print(f"Predicted Price for New Address: ${predicted_price[0][0]:.2f}")
```
